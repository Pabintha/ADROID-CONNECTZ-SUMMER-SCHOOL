```python
from itertools import combinations

def find_zero_sum_subsets(numbers):
    subsets = []
    num_elements = 5

    # Generate all possible combinations of 5 elements
    combinations_list = list(combinations(numbers, num_elements))

    # Check if the sum of each combination is zero
    for combination in combinations_list:
        if sum(combination) == 0:
            subsets.append(list(combination))

    return subsets

# Set of numbers
numbers = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]

# Find subsets with a sum of zero
zero_sum_subsets = find_zero_sum_subsets(numbers)

# Print the subsets
for subset in zero_sum_subsets:
    print(subset)
```

```python
from itertools import combinations

def find_zero_sum_subsets(numbers):
    subsets = []
    subset_sizes = range(3, 7)  # Subset sizes from 3 to 6

    # Generate all possible combinations of subset sizes
    for subset_size in subset_sizes:
        for combination in combinations(numbers, subset_size):
            if sum(combination) == 0:
                subsets.append(list(combination))

    return subsets

# Set of numbers
numbers = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]

# Find subsets with a sum of zero
zero_sum_subsets = find_zero_sum_subsets(numbers)

# Print the subsets
for subset in zero_sum_subsets:
    print(subset)
```
